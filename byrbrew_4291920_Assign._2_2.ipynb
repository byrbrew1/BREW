```python
#####Step 1: Import Libraries
#numpy for numerical operations.
#matplotlib.pyplot for plotting.
#sklearn.metrics.mean_squared_error to calculate the mean squared error.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
#####Step 2: Define Observations and Model Predictions
#observations: the actual observed temperatures.
#model_output: the predicted temperatures from a model.
observations = np.array([15.2, 16.1, 14.5, 15.8, 25.0]) 
model_output = np.array([14.8, 15.5, 14.0, 16.0, 25.3])
```

```python
#####Step 3: Define Error Variances
#sigma_o2 is the observation error variance (0.5 squared).
#sigma_b2 is the model background error variance (1.0 squared).
sigma_o2 = 0.5**2
sigma_b2 = 1.0**2
```

```python
#####Step 4: Calculate Weights
#weight_o is the weight for the observations.
#weight_b is the weight for the model output.
weight_o = sigma_b2 / (sigma_o2 + sigma_b2)
weight_b = sigma_o2 / (sigma_o2 + sigma_b2)
```

```python
#####Step 5: Perform Optimal Interpolation
analysis = weight_o * observations + weight_b * model_output
```

```python
#####Step 6: Calculate RMSE
rmse_model = np.sqrt(mean_squared_error(observations, model_output))
rmse_analysis = np.sqrt(mean_squared_error(observations, analysis))
```

```python
#####Step 7: Calculate Bias
bias_model = np.mean(model_output - observations)
bias_analysis = np.mean(analysis - observations)
```

```python
#####Step 8: Display Results
#Print the RMSE and bias for both the model output and the analysis.
print(f"RMSE (Model): {rmse_model}")
print(f"RMSE (Analysis): {rmse_analysis}")
print(f"Bias (Model): {bias_model}")
print(f"Bias (Analysis): {bias_analysis}")
```

```python
#####Step 9: Plotting the Results
#Plot the observations, model output, and analysis over the specified time steps.
#Create a figure with a specified size.
#Plot the observations with blue circles.
#Plot the model output with green squares.
#Plot the analysis with red stars.
#Label the x-axis and y-axis.
#Add a title to the plot.
#Add a legend to distinguish the lines.
#Enable the grid for better readability.
time_steps = [1, 2, 3, 4, 5]

plt.figure(figsize=(10, 5))
plt.plot(time_steps, observations, 'o-', label='Observations', markersize=10, color='blue')
plt.plot(time_steps, model_output, 's-', label='Model Output', markersize=10, color='green')
plt.plot(time_steps, analysis, '*-', label='Analysis', markersize=10, color='red')
plt.xlabel('TIME STEP (Days)', fontweight='bold', color='black')
plt.ylabel('TEMPERATURE (°C)', fontweight='bold', color='black')
plt.title('OBSERVATIONS, MODEL OUTPUT, & ANALYSIS OVER 5 DAYS', fontweight='bold', color='black')
plt.legend()
plt.grid(True)
```
